# Activity profiles for predicting depression
This is the Kaggle Depression Dataset containing actimetry data collected from patients suffering from Major Depression (MD) / Bipolar Disorder (BP) and controls (Ctrl) using a wearable actimeter (see: https://www.kaggle.com/datasets/arashnic/the-depression-dataset). MD is characterized by by feelings of sadness, helplessness, and lonliness, loss of drive and interest in daily activities, anxiety, as well as disrupted circadian rhythms and sleep disturbances. Patients suffering from BP additionally cycle between depressive and manic phases, with the latter being characterized by exaggerated mood (euphoria), drive (e.g. sexual drive), activity, and risky behavior.

Actimetry time series data is stored as separate csv files for each participant stored in two different folders for MD/BP patients and Ctrl, respectively. Additionally, general patient information collected during clinical examination has been collected in 'scores' csv file, containing information about clinical diagnosis and clinician-ratings using the 
Montgomery–Åsberg Depression Rating Scale (MADRS).

## load actimetry data
Actimeter data is stored as time series data in separate csv files for each subject. Each file contains a 'timestamp', 'date', and an 'activity' column.

```python
# import modules
import numpy as np
import pandas as pd
import os as os
import glob as glob
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns
from scipy import stats

%matplotlib inline
```

```python
# actimetry data are stored separately as individual csv files for each subject
# get a list of those files
wdir = os.getcwd()

file_list = []
for folder in ['condition', 'control']:
    for file in glob.glob(wdir + f'\\data\\{folder}\\*csv'): # use glob to extract file paths (w/ '\\') and append to list
        file_list.append(file)
```

## Explore and visualize data
Subject-specific actimeter data is stored in separate dataframes as a time series over several days, with data points recorded every minute.
Data is visualized for daily activity profiles for each subject to screen for patterns in activity and detect anomalies in the data.

```python
# define utility function for plotting daily activity patterns of individual subjects for all recording days
def plot_daily_activity(dataset, filename, grouping='date'):
    """"
    This function takes in the following input arguments: a dataset dictionary storing individual pd dataframes, 
    a filename, serving as a key to identifier the subject-specific dataframe in the dictionary, and a grouping variable, 
    set to 'date' by default to group time series data based on days.
    """
    # generate grouped dataframe
    grouped_df = dataset[filename].groupby(grouping)
    
    # initialize figure and set subplot arrangement
    fig = plt.figure(figsize=(10, 16))
    # n_cols is always 3, n_rows is number of plots / n_cols if there is no remainder, else add 1 more row
    col = 3
    row = len(grouped_df)//3 if len(grouped_df) % 3 == 0 else len(grouped_df)//3 + 1
    # initialze i for adding consecutive subplots
    i = 1
    #iterate over key-value pairs in grouped df and plot as subplot
    for key, data in grouped_df:
        ax = fig.add_subplot(row, col, i)
        # check if index is DatetimeIndex and define axis format for time series if True
        if isinstance(data.index, pd.DatetimeIndex):
            xtick_pos = mdates.HourLocator(byhour=range(0, 24, 6))
            xtick_fmt = mdates.DateFormatter('%H:%M')
            ax.xaxis.set_major_locator(xtick_pos)
            ax.xaxis.set_major_formatter(xtick_fmt)
        # for cases where index is e.g. type integer ranging between 0 and 24
        else:
            ax.set_xticks(np.arange(0, 25, 6)) # set to 25 (w/ stepsize=6) to also include 24 on x axis
        # plot data
        ax.plot(data.index, data.activity, label=key)
        ax.legend(loc = 'upper left')
        ax.set_xlabel('Time [h]')
        ax.set_ylabel('Activity [g/min]')
        i += 1
    plt.suptitle(filename)
    plt.tight_layout()
    plt.show()
```

## Clean and pre-process data
Exclude recording days with almost no activity and further pre-process the data.

```python
# initialize empty dataframe with columns for different activity thresholds
activity_counts = pd.DataFrame({'act_<20': np.zeros(len(filenames)), 
                                'act_<30': np.zeros(len(filenames)), 
                                'act_>30': np.zeros(len(filenames))},
                              index = filenames)
```

```python
#iterate over empty dataframe and compute value counts (in percent) for different thresholds per subject (index)
for key, value in actimeter_dict.items():
    grouped_df = actimeter_dict[key].groupby('date').mean('activity')
    activity_counts.loc[key, 'act_<20'] = round(sum(grouped_df['activity'] < 20)/len(grouped_df)*100, 2)
    activity_counts.loc[key, 'act_<30'] = round(sum(grouped_df['activity'] < 30)/len(grouped_df)*100, 2)
    activity_counts.loc[key, 'act_>30'] = round(sum(grouped_df['activity'] > 30)/len(grouped_df)*100, 2)
```

```python
activity_counts
```

Most of the low-activity days are identified using a threshold = 30 without excluding too many days. Therefore, exclusion threshold is set to 50 in order to avoid excluding to many days.

```python
# set threshold
thresh = 30

actimeter_cleaned = {}
for key, data in actimeter_dict.items():
    # extract all dates where activity > thresh and use to filter data
    dates = data.groupby('date')['activity'].mean().index[data.groupby('date')['activity'].mean() > thresh]
    actimeter_cleaned[key] = data[data['date'].isin(dates)]
```

```python
# plot cleaned recording days for subject 'condition_2'
plot_daily_activity(actimeter_cleaned, 'condition_2')
```

```python
# plot cleaned recording days for subject 'control_13'
plot_daily_activity(actimeter_cleaned, 'control_13')
```

```python
# average over daily hours to smoothen the data
actimeter_avg = {}
for key, data in actimeter_cleaned.items():
    data['hour'] = data.index.hour
    actimeter_avg[key] = data.groupby(['date', 'hour']).mean('activity').reset_index('date')
```

```python
# plot smoothened daily activity for subject 'condition_2'
plot_daily_activity(actimeter_avg, 'condition_2')
```

```python
# plot smoothened daily activity for subject 'control_13'
plot_daily_activity(actimeter_avg, 'control_13')
```

```python
# replace hourly activity by NaN if no activity was recorded for all days where number of hours with zero activity > 2
for key, data in actimeter_avg.items():
    # create new activity column with zeros to be replaed by NaN
    actimeter_avg[key]['activity_cleaned'] = data['activity']
    
    # select hours w/ zero activity, count those hours per date, and filter dates w/ zero hours > 2
    count = data[data['activity']==0].groupby('date').count()
    dates = count.index[count['activity']>2]
    
    # replace zeros w/ NaN for all days w/ zero hours > 2
    actimeter_avg[key].loc[actimeter_avg[key]['date'].isin(dates), 'activity_cleaned'] = data[data['date'].isin(dates)]['activity'].replace(0, np.nan)
```

## Feature engineering and data analysis
MD is characterized by overall reduced motivation and activity levels. Thus, the overall activity over days can be a valuable readout to compare subjects. In case of BP, patients additionally alternate between deppressive and hypo-/manic episodes, suggesting that variability between days, using coefficient of variation (CV), maybe used as a readout for cylcing between those states. Moreover, the day with minimum activity was computed as another readout for depression severity.

Patients may further show distorted circadian rhythms. Thus, in addition to those daily parameters, it may be valuable to calculate the mean activitx and CV for defined daytime periods, e.g. morning, afternoon, evening, night.

```python
# initialize a pd dataframe and assign group value to each subject
features = pd.DataFrame({'group': np.zeros(len(filenames))}, index=filenames)
for file in filenames:
    if 'condition' in file:
        features.loc[file, 'group'] = 1
    else:
        features.loc[file, 'group'] = 0
```

### Compute new features

```python
# compute included recording days, mean activity, and CV of activity, and append to dataframe using list comprehension
features['actdays_incl'] = [len(actimeter_avg[file].groupby('date')) for file in filenames] # used for quality control
features['mean'] = [actimeter_avg[file]['activity_cleaned'].mean() for file in filenames]
features['CV'] = [float(actimeter_avg[file].groupby('date')['activity_cleaned'].mean().std()) 
                  / actimeter_avg[file]['activity_cleaned'].mean() for file in filenames]
```

```python
# extract day with minimum activity, but only consider full-length days w/o NaN values to get an unbiased readout
# e.g. by excluding days w/ recordings only in the morning where activity levels are usually lower
for key, data in actimeter_avg.items():
    # extract valid, full-length days w/o NaN values  
    val_dates = data.groupby('date')['activity_cleaned'].count()[data.groupby('date')['activity_cleaned'].count()==24].index
    # filter for valid days, compute mean daily activity, extract minimum and append to dataframe
    features.loc[key, 'min'] = data[data['date'].isin(val_dates)].groupby('date')['activity_cleaned'].mean().min()
```

```python
features.head()
```

```python
# extract means and CVs for every hour and append to dataframe using list comprehension
# for every hour in a 24h interval, create new feature column and filled with mean activity for than hour (indexed by i)
for i in range(0, 24):
    features[f'mean_{str(i)}h'] = [actimeter_avg[file].loc[i,'activity_cleaned'].mean() for file in filenames]

for i in range(0, 24):
    features[f'CV_{str(i)}h'] = [float(actimeter_avg[file].loc[i, 'activity_cleaned'].std()) 
                                 / actimeter_avg[file].loc[i,'activity_cleaned'].mean() for file in filenames]
```

### Visualize and analyze data

```python
# define utility function for running one-way ANOVAs
def perform_oneway_anova(df, dv, iv):
    '''
    This function runs one-way ANOVA on input data and additionally tests ANOVA preconditions, i.e. normal distribution
    and homogeneity of variance. In case of significant group effect, pairwise post-hoc Tukey-HSD tests are performed:
    Input arguments: df (dataframe), dv (the dependent variable), iv (the independent variable)
    '''
    # test normal distribution and homogeneity of variance as ANOVA pre-conditions
    groups = [data for group, data in df.groupby(iv)[dv]] # get sublists of data grouped by iv 

    print('Shapiro-Wilk Test:')
    for group, data in df.groupby(iv):
        p_shapiro = stats.shapiro(data[dv])[1] # run test and extract p-value
        print(group, ' : p =', round(p_shapiro, 3), ' n.s.' if p_shapiro > 0.05 else '*')

    print('\nLevene Test:')
    p_levene = stats.levene(*groups)[1] # run test and extract p-value
    print('p =', round(p_levene, 3), ' n.s.' if p_levene > 0.05 else '*')
    
    # perform one-way ANOVA
    print('\nANOVA:')
    anova_res = stats.f_oneway(*groups) # run test and store in results variable
    print(f'F = {str(round(anova_res[0], 1))}', f', p = {str(round(anova_res[1], 3))}' # print F statistic and p-value
          if anova_res[1] > 0.001 else 'p < 0.001', ' n.s.' if anova_res[1] > 0.05 else '*')
    
    # perform pairwise post-hoc Tukey-HSD tests if ANOVA returns significant main effect
    if anova_res[1] < 0.05:
        print('\n Pairwise post-hoc Tukey-HSD Tests:')
        from statsmodels.stats import multicomp
        print(multicomp.pairwise_tukeyhsd(df[dv], df[iv], 0.05))
    else:
        None
```

## Disease prediction from actimetry data
After preprocessing and feature engineering, actimetry-generated features will be used to predict the presentce of MD and BP compared to Ctrl. Three main actimetry-based features were generated: 1) mean activity over days, 2) minimum activity, and 3) CV of activity between days. Since mean and minum activity were strongly correlated (r > 0.7) and reproduced similar group differences and minimum activity was additionally significantly correlated with the CV of activity (r > 0.5), minimum activity will be excluded from the feature set to avoid correlated features.

Since only 15 MD and 8 BP patients are contained in the dataset, making proper splits in training and test sets difficult, and no differences were observed between these patients subgroups, the logistic regression model will be trained on predicting the combined presence of MD and BP, hereinafter referred to as 'affective disorder' (AD).

### Train-Test Split

```python
# split into training and test set
from sklearn.model_selection import train_test_split

X = features[['mean', 'CV']]
y = features['group']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print(f"Training set: {X_train.shape} for X and {y_train.shape} for y.")
print(f"Test set: {X_test.shape} for X and {y_test.shape} for y.")

print(f"\nClass distribution in training set:\n{y_train.value_counts()}")
print(f"\nClass distribution in test set:\n{y_test.value_counts()}")
```

### Standardize features to z-scale

```python
#standardize X_train and project onto X_test
from sklearn.preprocessing import StandardScaler
z_scale = StandardScaler().fit(X_train)
X_train = z_scale.transform(X_train)
X_test = z_scale.transform(X_test)

#print means and SDs for X_train
print('Means and SDs for normalized X_train')
for column in range(X_train.shape[1]):
    print(round(np.mean(X_train[:, column]),2), ", ", round(np.std(X_train[:, column]),2),)

#print means and SDs for X_test
print('\nMeans and SDs for normalized X_test')
for column in range(X_test.shape[1]):
    print(round(np.mean(X_test[:, column]),2), ", ", round(np.std(X_test[:, column]),2))
```

### Logistic Regression

```python
# perform logistic regression
from sklearn.linear_model import LogisticRegression

# train and predict
lr = LogisticRegression(C=0.01, solver='liblinear').fit(X_train, y_train)
yhat = lr.predict(X_test)
```

### Determine accuracy on test set

```python
# import metrics for comparing model accuracy for models trained on original and reduced feature set
from sklearn.metrics import classification_report, confusion_matrix

# define utility function for confusion matrix
def plot_conf_matrix(conf_matrix, class_labels, title):
    """
    This function takes the confusion matrix computed using sklearn and corresponding class labels as arguments 
    and plots the data as a color-coded heatmap.
    """
    # normalize class predictions, returns percentage of class predictions for each label
    normalized = 100*(conf_matrix / np.sum(conf_matrix, axis = 1)[:, np.newaxis]) #sums over all column values in a row
    
    # plot data as heatmap using imshow
    fig, ax = plt.subplots(figsize=(4, 4))
    heatmap = ax.imshow(normalized, cmap='Blues')
    cbar = plt.colorbar(heatmap, shrink=0.8)
    cbar.set_label("% predicted", size=12)
    
    ax.set_title(f"{title}\n", size=16)
    tick_marks = range(len(class_labels)) # get number of tick marks
    ax.set_xticks(tick_marks, class_labels, size=12, rotation=45) # plot corresponding class labels
    ax.set_yticks(tick_marks, class_labels, size=12) # ""
    ax.set_xlabel("Predicted", size=14)
    ax.set_ylabel("True", size=14)
    
    #loop over range of class labels in 2D (as matrix coordinates) and plot data as cell values
    for i in range(len(class_labels)):
        for j in range(len(class_labels)):
            ax.text(y=i, x=j,
                    #plot percent predicted (i.e. normalized data), followed by absolute number (in square brackets) 
                    s=f"{round(normalized[i,j], 1)}%\n[{conf_matrix[i,j]}]",
                    horizontalalignment='center', verticalalignment='center',
                    #if value > average (resulting in dark cell color), use white as font color, else black
                    color="white" if conf_matrix[i, j] > conf_matrix.mean() else "black", size=11)
```

```python
#classification report
print('Model accuary\n\n', classification_report(y_test, yhat))

#compute confusion matrix and plot
cm = confusion_matrix(y_test, yhat, labels=[0, 1])
plot_conf_matrix(cm, ['Ctrl', 'AD'], 'Confusion Matrix')
```
